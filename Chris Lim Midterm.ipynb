```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from yellowbrick.regressor import ResidualsPlot
from sklearn.model_selection import cross_val_score
# H. Build a linear regression model with mpg as the target and displacement as the predictor. (3 marks)
print("Part H")
x = df.displacement.values
x = x.reshape(-1,1)
y = df.mpg.values
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
#print(X_train)
#print(y_train)
regressor.fit(X_train, y_train)
# a. For your model, what is the value of the intercept β0 ? (1 marks)
print("a,")
print("The intercept is", regressor.intercept_)
print("")
# b. For your model, what is the value of the coefficient β1 of the attribute displacement? (1 marks)
print("b,")
print("The coefficient is", regressor.coef_)
print("")
# c. What is the regression equation as per the model? (2 marks)
print("c,")
print("y (mpg) = ", regressor.coef_, "x (displacement) + ", regressor.intercept_)
print("")
# d. For your model, does the predicted value for mpg increase or decrease as the displacement increases? (1 marks)
print("d,")
print("Due to the negative slope, mpg decreases as the displacement increases")
print("")
# e. Given a car with a displacement value of 220, what would your model predict its mpg to be? (2 marks)
y_pred = regressor.predict([[220]])
print("e,")
print("We would predict the mpg to be", y_pred)
print("")
# f. Display a scatterplot of the actual mpg vs displacement and superimpose the linear regression line. (2 marks)
print("f,")
import matplotlib.pyplot as plt
import numpy as np
x_range = np.linspace(min(X_train),max(X_train)) # generate all possible Xs
x_range = x_range.reshape(-1,1)
y_predictions = regressor.predict(x_range)
plt.scatter(X_train, y_train, color='blue')
plt.plot(x_range,y_predictions,color='red',linewidth=3)
plt.xlabel("displacement")
plt.ylabel("mpg")
plt.title('Displacement vs mpg')
plt.show()
print("")
# g. Plot the residuals. (2 marks)
print("g,")
y_test_pred = regressor.predict(X_test)
print("R^2: {}".format(regressor.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_test_pred))
print("Root Mean Squared Error: {}".format(rmse))
visualizer = ResidualsPlot(regressor)
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.poof();
print("Blue is the training data, green is the test data")

```
